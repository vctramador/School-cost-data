```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Tuition_Assistance_20240525.csv")
df
```

```python
dados_sem_nan = df.dropna()
dados_sem_nan = df.dropna(axis=0)
```

```python
school_exp_idx = df['Cost'].idxmax()
school_exp = df.iloc[school_exp_idx]

school_name = school_exp['School']
school_cost = school_exp['Cost']
print(f"Escola com maior custo: {school_name} - R$ {school_cost:.2f}")
```

```python
#showing the expensives schools (without NaNs values)
no_nan = df.dropna()
no_nan = df.dropna(axis=0)
df_ordenado = no_nan.sort_values(by='Cost', ascending=False)

top_10 = df_ordenado.head(10)
top_10
```

```python
#showing de cheaper schools (without NaNs values)
filtro_df = no_nan[no_nan['Cost'] > 0]
df_ordenado = filtro_df.sort_values(by='Cost', ascending=True)


top_10_cheap = df_ordenado.head(10)
top_10_cheap
```

```python
mean = no_nan['Cost'].mean()
mean
```

```python
#most expensives schools (by mean)
mean_costs = df.groupby('School')['Cost'].mean()
sorted_mean_costs = mean_costs.sort_values(ascending=False)
sorted_mean_costs
```

```python
#mean cost by department
cost_by_department = df.groupby('Department')['Cost'].mean().reset_index()
cost_by_department1 = cost_by_department.sort_values(by='Cost', ascending=False)
cost_by_department1
```

```python
#total cost by school
cost_per_school = df.groupby('School')['Cost'].sum().reset_index()
cost_per_school = cost_per_school.sort_values(by='Cost', ascending=False)
cost_per_school
```

```python
#number of courses by school
courses_numbers_by_school = df.groupby('School')['Course Title'].count().reset_index()
courses_numbers_by_school = courses_numbers_by_school.rename(columns={'Course Title': 'Course Count'})
courses_numbers_by_school = courses_numbers_by_school.sort_values(by='Course Count', ascending=False)
courses_numbers_by_school
```
